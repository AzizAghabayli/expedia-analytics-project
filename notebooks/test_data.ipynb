```python
import pandas as pd
```

```python
test_df = pd.read_csv('../data/raw/test.csv')
test_df.head()
```

```python
num_rows, num_columns = test_df.shape
print(f"Number of columns in test.csv: {num_columns}")
print(f"Number of rows in test.csv: {num_rows}")
```

```python
# Listing column names for test.csv DataFrame
test_column_names = test_df.columns.tolist()
print("Column names in test.csv:", test_column_names)
```

```python
# Creating a DataFrame for column information
test_columns_info = pd.DataFrame({
    'Column Name': test_df.columns,
    'Data Type': [test_df[col].dtype for col in test_df.columns]
})

# Display the DataFrame
test_columns_info
```

```python
test_columns_info.to_csv('test_columns_info.csv', index=False)
print('Dataframe exported to csv file.')
```

```python
test_columns_info.to_csv('test_columns_info.txt', index=False)
print('Dataframe exported to txt file.')
```

```python
# Select the first few rows. For example, the first 5 rows
test_col_head = test_df.head(5)

# Export to CSV
test_col_head.to_csv('test_col_head.csv', index=False)
test_col_head.to_csv('test_col_head.txt', index=False)
```
