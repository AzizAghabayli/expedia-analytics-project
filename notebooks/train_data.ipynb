```python
import pandas as pd
```

```python
train_df = pd.read_csv('../data/raw/train.csv')
train_df.head()
```

```python
num_rows, num_columns = train_df.shape
print(f"Number of columns in train.csv: {num_columns}")
print(f"Number of rows in train.csv: {num_rows}")
```

```python
# Listing column names for train.csv DataFrame
test_column_names = train_df.columns.tolist()
print("Column names in train.csv:", test_column_names)
```

```python
# Creating a DataFrame for column information
train_columns_info = pd.DataFrame({
    'Column Name': train_df.columns,
    'Data Type': [train_df[col].dtype for col in train_df.columns]
})

# Display the DataFrame
train_columns_info
```

```python
train_columns_info.to_csv('train_columns_info.csv', index=False)
print('Dataframe exported to csv file.')
```

```python
train_columns_info.to_csv('train_columns_info.txt', index=False)
print('Dataframe exported to txt file.')
```
